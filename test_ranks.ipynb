```python
import pandas as pd
```

```python
events = pd.read_parquet(path='./data/events.parquet')
```

```python
user_id = 661379
user_items = events[events['user_id'] == user_id][['user_id','item_id']]
```

```python
items = pd.read_parquet(path='./data/items.parquet')
```

```python
for_agg = ['artists','albums','genres']
user_items =user_items.merge(items[['item_id'] + for_agg],on='item_id')
```

```python
user_items
```

```python
weights = {}
for fild in for_agg:
    calc_weight = user_items.explode(fild).groupby(fild).agg(score=(fild,"count"))
    calc_weight['score'] = calc_weight['score'] / calc_weight['score'].sum()
    weights[fild] = calc_weight

weights
```

```python
item_ids = [44279380, 51516485, 28036311, 36664892, 44484514, 51241318, 34568038, 44184116, 45499814, 41667460]
items_for_rank = items[items['item_id'].isin(item_ids)]
items_for_rank
```

```python
items_for_rank['score_total'] = 0
for fild in for_agg:
    if weights[fild] is not None:
        target_indexes = list(weights[fild].index)
        items_for_rank['score_' + fild] = [weights[fild].loc[x].sum()['score'] if set(x).issubset(target_indexes) else 0 for x in items_for_rank[fild]]
        items_for_rank['score_total'] += items_for_rank['score_' + fild]

recs_df = items_for_rank.sort_values(by='score_total',ascending=False)
recs_df
```
